```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df['floors'])
plt.pyplot.xlabel('floors')
plt.pyplot.ylabel('count')
plt.pyplot.title('floor amount')
```

```python
bins = np.linspace(min(df['floors']), max(df['floors']), 4)
```

```python
group_names = ['Low', 'Medium', 'High']
```

```python
df['floors-binned'] = pd.cut(df['floors'], bins, labels=group_names, include_lowest=True)
df[['floors', 'floors-binned']].head(20)
```

```python
df['floors-binned'].value_counts()
```

```python
pyplot.bar(group_names, df['floors-binned'].value_counts())
plt.pyplot.xlabel('floors')
plt.pyplot.ylabel('count')
plt.pyplot.title('floor-bins amount')
```
